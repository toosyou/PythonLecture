```python
import numpy as np
import matplotlib.pyplot as plt

# to plot figures in the notebook
%matplotlib inline
```

# Generate fake signals

```python
t = np.arange(0, 10, 0.01)

t[:10]
```

```python
signal = np.sin(t)
signal2 = np.cos(t)

signal[:10], signal2[:10]
```

# `plt.plot`

### If we only want to plot signal

```python
plt.figure(figsize=(20, 3)) # to make the figure bigger
plt.plot(signal)
```

### If we want both time and signal

```python
plt.figure(figsize=(20, 3)) # to make the figure bigger
plt.plot(t, signal)
```

### If we want both signal in the same plot

```python
plt.figure(figsize=(20, 3)) # to make the figure bigger
plt.plot(t, signal)
plt.plot(t, signal2)
```

### If we want both plots in the same figure
* We can utilize `plt.subplots` which will return a `Figure` and a list of `Axes`
    * We can then plot using `Axes`

```python
nrows, ncols = 2, 1
fig, axes = plt.subplots(nrows, ncols, figsize=(20, 6))
axes[0].plot(t, signal)
axes[1].plot(t, signal2)

# optional
fig.suptitle('Signals') # title of the whole plot

axes[0].set_title('sin(t)') # titles of subplots
axes[1].set_title('cos(t)')
```

# `plt.hist` for histograms

```python
plt.figure(figsize=(15, 10))
plt.hist(signal, bins=20, alpha=0.3, label='signal')

# optional
plt.hist(signal2, bins=20, alpha=0.3, label='signal2') # second hist

plt.legend()
```

# `plt.scatter` for scatters

```python
random_indices = np.random.randint(signal.shape[0], size=30) # randomly select points

plt.figure(figsize=(20, 10))
plt.scatter(t[random_indices], signal[random_indices])
```

# `plt.pie` for pie charts

```python
n = [10, 30, 100]
name = ['blue pizza', 'orange pizza', 'green pizze']

plt.figure(figsize=(20, 10))
patches, texts = plt.pie(n)
plt.legend(patches, name) # for labels
```
